```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
df = pd.read_csv('data.csv')

table_data = pd.DataFrame({'Data Type': df.dtypes, 'Null Values': df.isnull().sum()})
print(table_data)
```

```python
data_apr = df.groupby(['order_id', 'item_name'])[['quantity']].sum().unstack().reset_index().fillna(0) 
data_apr.drop(columns=data_apr.columns[0],axis =1, inplace = True)
def num(x):
    x=int(x)
    if x <= 0: return 0
    elif x >=1: return 1
data_apr = data_apr.applymap(num)
data_apr = data_apr['quantity']
```

```python
s = 0.01
c = 0.01
frequent_itemsets = apriori(data_apr, min_support = s, use_colnames = True)
rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = c)
print(f'\n------------------------------------------------\nfor Support={s:.2f}\t\tConfidence={c:.2f}')
print("Frequent itemsets are: \n",frequent_itemsets.sort_values(by="support",ascending=False)[0:10])
print("\nAssociation Rules are: \n", rules[["antecedents","consequents","confidence"]].sort_values(by="confidence",ascending=False)[:])
```

```python
s=0.01
c=0.2
while s<=0.2:
    frequent_itemsets = apriori(data_apr, min_support = s, use_colnames = True)
    while c<0.5:
        rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = c)
        print(f'\n------------------------------------------------\nfor Support={s:.2f}\t\tConfidence={c:.2f}')
        print("Frequent itemsets are: \n",frequent_itemsets[0:2])
        print("\nAssociation Rules are: \n", rules[["antecedents","consequents","confidence"]][0:2].sort_values(by="confidence",ascending=False))

        print()
        c+=0.015
    s+=0.01
```
